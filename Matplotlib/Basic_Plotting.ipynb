### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# ploting a simple function

price = [48000,54000,57000,49000,45000]
year = [2015,2016,2017,2018,2019]

plt.plot(year,price)
```

```python
# from a pandas dataframe

batsman = pd.read_csv('D:\mca_aiml_lab\mca_env\datasets-session-23\sharma-kohli.csv')

batsman.head()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plot

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('V.Kholi vs Rohit Sharma Career Comparion in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# color(hex) and line(width and style) and markers(size)

plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2)

plt.title('V.Kholi vs Rohit Sharma Career Comparion in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# marker 

plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=3,marker='D',markersize=7)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2,marker='+',markersize=7)

plt.title('V.Kholi vs Rohit Sharma Career Comparion in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# legend -> axes

plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=3,marker='D',markersize=7,label='VKohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2,marker='+',markersize=7,label='RSharma')

plt.title('V.Kholi vs Rohit Sharma Career Comparion in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')

plt.legend(loc='best')
```

```python
# limiting axes

price = [48000,54000,57000,49000,45000,4500000]
year = [2015,2016,2017,2018,2019,2021]

plt.plot(year,price)
plt.ylim(0,75000)
```

```python
# grid and show


plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=3,marker='D',markersize=7,label='VKohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2,marker='+',markersize=7,label='RSharma')

plt.title('V.Kholi vs Rohit Sharma Career Comparion in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')

plt.legend(loc='best')
plt.grid()
plt.show()
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,30,10,50]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart

plt.barh(colors,children,color='green')
```

```python
# multiple bar plot 

df = pd.read_csv(r'D:\mca_aiml_lab\mca_env\datasets-session-23\batsman_season_record.csv')

df
```

```python
# xticks
plt.bar(np.arange(df.shape[0]) -0.2,df['2015'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]) +0.2,df['2017'],width=0.2,color='orange')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# problem

children = [10,20,30,10,50]
colors = ['redredredredred','blueblueblueblueblue','greengreengreengreengreen','yellowyellowyellowyellowyellow','pinkpinkpinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')   # handle longest name
plt.show()
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2017'],label=2017)
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+ df['2017']),label='2015')

plt.legend()

plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data

data = [23,45,100,20,49]
subjects = ['eng','science','math','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset

cg = pd.read_csv('D:\mca_aiml_lab\mca_env\datasets-session-23\gayle-175.csv')

cg.head()


```

```python
plt.pie(cg['batsman_runs'],labels=cg['batsman'],autopct='%0.1f%%')

plt.show()
```

```python
# percentage and colors

plt.pie(cg['batsman_runs'],labels=cg['batsman'],autopct='%0.1f%%',colors=['blue','green','orange','red','pink','brown'])

plt.show()
```

```python
# Expload and shadow

plt.pie(cg['batsman_runs'],labels=cg['batsman'],autopct='%0.1f%%',explode=[0.2,0,0,0,0,0.1],shadow=True)


plt.show()

```
