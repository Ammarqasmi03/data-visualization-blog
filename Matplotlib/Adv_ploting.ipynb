```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax[1].scatter(batters['avg'],batters['runs'],color='green',marker='D')

ax[0].set_title('Avg vs Strike Rate Analysics')

ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg vs Batsman Runs Analysics')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')




```

```python
fig,ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax[0,1].scatter(batters['avg'],batters['runs'],color='green',marker='D')
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])


ax[0,0].set_title('Avg vs Strike Rate Analysics')
ax[0,0].set_ylabel('Strike Rate')

ax[0,1].set_title('Avg vs Batsman Runs Analysics')
ax[0,1].set_xlabel('Avg')
ax[0,1].set_ylabel('Runs')
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)  # (row,col,graph number)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['avg'])

ax2 = fig.add_subplot(2,2,3)
ax2.hist(batters['runs'])
```

### 3D surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx , yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2 
z.shape
```

```python
fig = plt.figure(figsize=(15,10))

ax = plt.subplot(projection ='3d')

p = ax.plot_surface(xx,yy,z,cmap='plasma')
fig.colorbar(p)

```

```python
fig = plt.figure(figsize=(15,10))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='plasma')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(15,10))

ax = plt.subplot(projection ='3d')

p = ax.plot_surface(xx,yy,z,cmap='plasma')
fig.colorbar(p)
```

#### Contour Plots

```python
fig = plt.figure(figsize=(15,10))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='plasma')
fig.colorbar(p)
```

#### HeapMap

```python
delivery = pd.read_csv('D:\mca_aiml_lab\mca_env\datasets-session-24\IPL_Ball_by_Ball_2008_2022.csv')

delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]

temp_df.head()
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')

grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
